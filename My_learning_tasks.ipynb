```python
job_counts = df['job_title'].nunique()
job_counts
```

####There are 125 unique job titles in our datset so it would be difficult to fit all in the dataset hence we inspect the data set further and se the count of the each job title

```python
job_counts_unique = df['job_title'].value_counts().head(10)
job_counts_unique
```

####We have selected the 10 most occuring job titles and we then proceed to draw the pie chart

```python
import matplotlib.pyplot as plt

top_eight = 8
top_job_counts=job_counts_unique.head(top_eight)
plt.pie(top_job_counts,labels=top_job_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("The most common job titles")
plt.axis=('equal')
plt.show()

```

####Now we select and create a new dataframe for only those individuals with Data Scientist as their job title and export this to form our new dataframe

##Question Five

```python
#Analysing salary by experience level
experience_level=new_df.groupby('experience_level')['salary'].agg(['sum', 'min', 'max', 'mean'])
experience_level
```

######The above function uses the .groupby function to group salary based on experience level from entry-levl to senior. As seen we can see that the salary increases based on levelo seniority but would be great to visualize it as done below

```python
experience_level = df.groupby('experience_level')['salary'].agg(['sum', 'min', 'max', 'mean'])

# Create individual bar charts for each aggregate statistic
for column in experience_level.columns:
    plt.figure(figsize=(8, 6))
    plt.bar(experience_level.index, experience_level[column], color='skyblue')
    plt.title(f'{column} by Experience Level')
    plt.xlabel('Experience Level')
    plt.ylabel(column)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

######Going by the above bar charts then we can conclude that the salary increases significantly by level of experience

#### Using a boxplot we will visualize compare the salaries by employment type, so we create a new dattaframe setting United Kingdom as the employee residence and move on to draw the plot


```python
new_df.head()
```

```python
new_df_uk=new_df[new_df['employee_residence']=='United Kingdom']
new_df_uk

salary_by_type_of_employment = new_df_uk.groupby('employment_type')['salary'].describe()


plt.figure(figsize=(10, 6))

plt.boxplot([
    new_df_uk[new_df_uk['employment_type'] == 'Full-time']['salary'],
    new_df_uk[new_df_uk['employment_type'] == 'Part-time']['salary'],
    new_df_uk[new_df_uk['employment_type'] == 'Contract']['salary']
], labels=['Full-time', 'Part-time', 'Contract'])
plt.title('Salary Comparison by Employment Type in the United Kingdom')
plt.suptitle('')
plt.xlabel('Employment Type')
plt.ylabel('Salary')
plt.xticks(rotation=45)

plt.show()
```
